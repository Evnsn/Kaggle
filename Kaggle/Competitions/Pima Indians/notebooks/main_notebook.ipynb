#### Visualize the data

```python
for column in originale_data.columns:
    plt.hist(originale_data[column])
    plt.title(column)
    plt.show()
```

```python
# Correlation (before or after transformation?)
sns.heatmap(originale_data.corr())
```

```python
# Density plot
for column in originale_data.columns:
    if column == 'Outcome':
        continue
    sns.displot(data=originale_data, x=column, hue="Outcome", kind="kde")
    plt.title(column + " vs Outcome")
    plt.show()
```
